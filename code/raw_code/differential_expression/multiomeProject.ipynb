```r
PROJDIR='/projects/pfenninggroup/singleCell/Pierce_transgen_rat_multiome/data/tidy_data/ArchRProjects'
projM = loadArchRProject(file.path(PROJDIR,'Rat_Transgen_NAc_multiome'), showLogo = FALSE)
```

```r
projM <- addMotifAnnotations(ArchRProj = projM, motifSet = "cisbp", name = "Motif", species = "mus musculus")
```

```r
projM <- addBgdPeaks(projM)
```

```r
projM <- addDeviationsMatrix(
  ArchRProj = projM, 
  peakAnnotation = "Motif",
  force = TRUE
)
```

```r
projM = saveArchRProject(ArchRProj = projM)
```

```r
getAvailableMatrices(ArchRProj = projM)
```

```r
seGroupMotif <- getGroupSE(ArchRProj = projM, useMatrix = "MotifMatrix", groupBy = "Sire")
```

```r
seGroupMotif
```

```r
seZ <- seGroupMotif[rowData(seGroupMotif)$seqnames=="z",]
```

```r
rowData(seZ)$maxDelta <- lapply(seq_len(ncol(seZ)), function(x){
  rowMaxs(assay(seZ) - assay(seZ)[,x])
}) %>% Reduce("cbind", .) %>% rowMaxs
```

```r
corGSM_MM_Harmony <- correlateMatrices(
    ArchRProj = projM,
    useMatrix1 = "GeneScoreMatrix",
    useMatrix2 = "MotifMatrix",
    reducedDims = "HarmonyI_Combined"
)
```

```r
corGSM_MM_Harmony$maxDelta <- rowData(seZ)[match(corGSM_MM_Harmony$MotifMatrix_name, rowData(seZ)$name), "maxDelta"]
```

```r
corGSM_MM_Harmony <- corGSM_MM_Harmony[order(abs(corGSM_MM_Harmony$cor), decreasing = TRUE), ]
corGSM_MM_Harmony <- corGSM_MM_Harmony[which(!duplicated(gsub("\\-.*","",corGSM_MM_Harmony[,"MotifMatrix_name"]))), ]
corGSM_MM_Harmony$TFRegulator <- "NO"
corGSM_MM_Harmony$TFRegulator[which(corGSM_MM_Harmony$cor > 0.5 & corGSM_MM_Harmony$padj < 0.01 & corGSM_MM_Harmony$maxDelta > quantile(corGSM_MM_Harmony$maxDelta, 0.75))] <- "YES"
sort(corGSM_MM_Harmony[corGSM_MM_Harmony$TFRegulator=="YES",1])
```

```r
p <- ggplot(data.frame(corGSM_MM_Harmony), aes(cor, maxDelta, color = TFRegulator)) +
  geom_point() + 
  theme_ArchR() +
  geom_vline(xintercept = 0, lty = "dashed") + 
  scale_color_manual(values = c("NO"="darkgrey", "YES"="firebrick3")) +
  xlab("Correlation To Gene Score") +
  ylab("Max TF Motif Delta") +
  scale_y_continuous(
    expand = c(0,0), 
    limits = c(0, max(corGSM_MM_Harmony$maxDelta)*1.05)
  )

p
```
